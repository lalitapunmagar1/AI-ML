#Why choose Matplotlib and Seaborn
Both libraries are among the most popular choices for data visualization in Python. Here's why:

Matplotlib: Known for its flexibility, Matplotlib allows detailed customization of visual elements, making it perfect for tailored plots.

Seaborn: Built on top of Matplotlib, Seaborn simplifies visualization tasks by offering high-level functions to create aesthetically pleasing statistical graphics.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Basics of matplotlib
#Line plot
x= np.linspace(0, 10, 100)
y= np.sin(x)
plt.plot(x,y)
```

```python
x= np.linspace(0, 10, 100)
y= np.sin(x)
plt.plot(x, y, label= "Sine wave")
plt.title("Line plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
#Bar chart
categories= ['A', 'B', 'C']
values= [10, 20, 30]
plt.bar(categories, values, color= ["blue", "red", "green"])
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

#Visualization  with seaborn

```python
import seaborn as sns
```

```python
#Histogram and KDE plot:
data= np.random.randn(1000)
sns.histplot(data, kde= True, color="black")
plt.title("Hostogram with KDE")
plt.show()
```

```python
# heatmap
matrix= np.random.rand(5, 5)
matrix
```

```python
sns.heatmap(matrix, annot= True, cmap= "vanimo")
plt.title("Heatmap Example")
plt.show()
```

```python
#pairplot example
sns.set_style("whitegrid")
df= sns.load_dataset("iris")
sns.pairplot(df, hue= "species", markers=["o", "s", "D"])
plt.show()
```

```python
import pandas as pd
import numpy as
df= pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Customer Call List.xlsx")
df
```

```python
df= df.drop_duplicates()
df
```

```python
print(df.columns)
```

```python
df= df.drop('Not_Useful_Column', axis=1)
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip()
df
```

```python
df['Last_Name'] = df['Last_Name'].str.title()
df
```

```python
df['Last_Name'] = df['Last_Name'].str.replace(r'[^\w\s]', '', regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].replace(['N/A', 'n/a', np.nan], '')
df
```

```python
df[['street_address', 'state', 'zip_code']] = df['Address'].str.split(',', expand=True)
df['street_address'] = df['street_address'].str.strip()
df['state'] = df['state'].str.strip()
df['zip_code'] = df['zip_code'].str.strip()
df
```
